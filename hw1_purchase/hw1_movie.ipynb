# (2) The actor generating the highest average revenue?

```python
df2actor = df["Actors"].str.split("|",expand = True,n = None)#因為一部電影有4位演員塞在同一個string裡面，所以用pandas內的.str.split()拆分"|"，且因為需要把每個人名獨立出來而不是放在一個list內，所以用"expand = True"來進行DataFrame expanding dimensionality，n=none來表示所有的"|"都要拆除
df2actor = df2actor.stack(level = -1,dropna = False)#用stack()來將從同一個row的演員拆開後放在同個column，dropna=False為不要刪除缺失值
df2actor = df2actor.reset_index(level = 1,drop = True,inplace = False)#重新設定index，level = 1為刪去第1行，保留第0行的index，drop=True為了不讓舊的index被添加的新的column裡面，inplace = False為就地修改dataframe而不去創造一個新的object
df2actors = df2actor.to_frame(name='Actors')#因為df2actors是一個series，我想把它變成一個dataframe(為了要跟revenue合併)
df2 = df[["Revenue (Millions)"]]#創造只有revenue的dataframe
df2 = df2.merge(df2actors, how ='inner', left_index = True, right_index = True,sort = None)#將兩個dataframe合併,on = None且不合併索引，則默認為兩個DataFrame中columns的交集，left_index=True以及right_index=True，使用來自左右DataFrame的index作為連接鍵，sort = None表示不特別進行排序。
df2 = df2.groupby("Actors",axis = 0,observed = False) #因為要算每個演員的薪資平均，所以用groupby函數讓演員變成每個row的index，axis = 0是要切分時是用Row來進行而不是column,dropna = True是要讓缺失值刪掉，observed = False用意在顯示categorical groupers的所有值
df2
```

# (4) Top‐3 directors who collaborate with the most actors? 

```python
df4_directors = df["Director"].str.strip() #把每一個Director做處理如果前後有空格的話去掉空格
df4_directors = df4_directors.to_frame() #把這個series變成dataframe
df4_actor = df["Actors"].str.strip().str.split("|",expand = True,n = None) #把每個actors做處理，去掉空格並且以|來拆分
df4_actor = df4_actor.stack(level = -1,dropna = False) #同個row的堆疊起來放在一起
df4_actor = df4_actor.reset_index(level = 1,drop = True,inplace = False) #重新設定一下index
df4_actor = df4_actor.to_frame(name='Actors') #把actor這個series變成dataframe
df4 = df4_directors.merge(df4_actor, how ='inner', left_index = True, right_index = True,sort = None) #合併兩個dataframe
df4 = df4.groupby("Director",axis = 0,observed = False) #用groupby把導演變成index(這裡用第一次groupby是為了要讓導演變成index然後刪去重複的演員)
df4 = df4.apply(lambda x: x.drop_duplicates('Actors')) #用apply以及lambda函數把同個導演重複的actor去掉
df4 = df4.rename(columns={'Director':'Directors'},inplace = False ,level = None) #此時index_col以及第0個column都是director，所以用directors做區分
df4 = df4.groupby(['Directors']).count() #再做一次groupby並計算同個導演的合作演員數量(這裡又用了一次groupby是為了計算出合作的演員數)
df4 = df4.sort_values(['Actors'],ascending=False) #排序
df4.head(n=3) #取前三
```

# (5) Top‐2 actors playing in the most genres of movies?

```python
df5actor = df["Actors"].str.split("|",expand = True,n = None) #把每個actors做處理，以|來拆分
df5genre = df["Genre"].str.split("|",expand = True,n = None) #把每個Genre做處理，以|來拆分
df5actor = df5actor.stack(level = -1,dropna = False) #同個row的堆疊起來放在一起
df5genre = df5genre.stack(level = -1,dropna = False) #同個row的堆疊起來放在一起
df5genre = df5genre.reset_index(level = 1,drop = True,inplace = False) #重新設定一下index
df5actor = df5actor.reset_index(level = 1,drop = True,inplace = False) #重新設定一下index
df5actor = df5actor.to_frame() #把這個series變成dataframe
df5genre = df5genre.to_frame() #把這個series變成dataframe
df5 = df5actor.merge(df5genre, how ='inner', left_index = True, right_index = True,sort = None) #合併兩個dataframe
df5 = df5.rename(columns={'0_x':'Actors','0_y':'Genre'},inplace = False ,level = None ) #在此dataframe中兩個剛合併的資料column name分別是'0_x'跟'0_y'要改成'Actors'跟'Genre'
df5["Genre"] = df5["Genre"].str.strip() #把每個Genre做處理，去掉空格
df5["Actors"] = df5["Actors"].str.strip() #把每個actors做處理，去掉空格
df5 = df5.groupby("Actors",axis = 0,observed = False)  #用groupby把Actors變成index(這裡用第一次groupby是為了要讓Actors變成index然後刪去重複的演員)
df5 = df5.apply(lambda x: x.drop_duplicates('Genre')) #用apply以及lambda函數把同個Actors重複的Genre去掉
df5 = df5.rename(columns={'Actors':'Actor'},inplace = False ,level = None) #此時index_col以及第0個column都是Actors，所以用Actor做區分
df5= df5.groupby(['Actor']).count() #再做一次groupby並計算同個Actor的Genre數量(這裡又用了一次groupby是為了計算出Genre數)
df5 = df5.sort_values(['Genre'],ascending=False) #排序
df5.head(n=6) #取前二(第二名有5個)
```

# (6)Top‐3 actors whose movies lead to the largest maximum gap of years?

```python
df6actor = df["Actors"].str.split("|",expand = True,n = None) #把每個actors做處理，以|來拆分
df6year = df["Year"]
df6actor = df6actor.stack(level = -1,dropna = False) #同個row的堆疊起來放在一起
df6actor = df6actor.reset_index(level = 1,drop = True,inplace = False) #重新設定一下index
df6actor = df6actor.to_frame() #把這個series變成dataframe
df6year = df6year.to_frame() #把這個series變成dataframe
df6 = df6actor.merge(df6year, how ='inner', left_index = True, right_index = True,sort = None) #合併兩個dataframe
df6 = df6.rename(columns={0:'Actors','0_y':'Year'},inplace = False ,level = None ) #在此dataframe中兩個剛合併的資料column name分別是'0_x'跟'0_y'要改成'Actors'跟'Genre'
df6 = df6.groupby("Actors",axis = 0,observed = False)  #用groupby把Actors變成index(這裡用第一次groupby是為了要讓Actors變成index然後刪去重複的演員)
df6 = df6.max()-df6.min()
df6 = df6.rename(columns={'Year':'Gap_Year'},inplace = False ,level = None ) #在此dataframe中兩個剛合併的資料column name分別是'0_x'跟'0_y'要改成'Actors'跟'Genre'
df6 = df6.sort_values(['Gap_Year'],ascending=False) #排序
df6.max()
df6[df6['Gap_Year'] == 10]
```
