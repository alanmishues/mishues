# crawler THSR information
Let's start to crawl the imformation of each station of THSR!
* Target web: [THSR Homepage](https://www.thsrc.com.tw)  
* Choose each station:
    1. Homepage -> Travel Information -> Stations
    2. choose different stations by changing url
* Task:
    1. crawler the name, address, operation hours and ticketing hours of each station
    2. organize into a DataFrame

```python
# setting each station name and their url
url_list =['https://www.thsrc.com.tw/ArticleContent/2f940836-cedc-41ef-8e28-c2336ac8fe68',
          'https://www.thsrc.com.tw/ArticleContent/977abb69-413a-4ccf-a109-0272c24fd490',
          'https://www.thsrc.com.tw/ArticleContent/e6e26e66-7dc1-458f-b2f3-71ce65fdc95f',
          'https://www.thsrc.com.tw/ArticleContent/fbd828d8-b1da-4b06-a3bd-680cdca4d2cd',
           'https://www.thsrc.com.tw/ArticleContent/a7a04c89-900b-4798-95a3-c01c455622f4',
          'https://www.thsrc.com.tw/ArticleContent/e8fc2123-2aaf-46ff-ad79-51d4002a1ef3',
          'https://www.thsrc.com.tw/ArticleContent/3301e395-46b8-47aa-aa37-139e15708779',
          'https://www.thsrc.com.tw/ArticleContent/38b8c40b-aef0-4d66-b257-da96ec51620e',
          'https://www.thsrc.com.tw/ArticleContent/5f4c7bb0-c676-4e39-8d3c-f12fc188ee5f',
          'https://www.thsrc.com.tw/ArticleContent/60831846-f0e4-47f6-9b5b-46323ebdcef7',
          'https://www.thsrc.com.tw/ArticleContent/9c5ac6ca-ec89-48f8-aab0-41b738cb1814',
          'https://www.thsrc.com.tw/ArticleContent/f2519629-5973-4d08-913b-479cce78a356']
station_list = ['南港','台北','板橋','桃園','新竹','苗栗','台中','彰化','雲林','嘉義','台南','左營']
```

```python
# crawler the address of each station
address_list = []
for i in url_list:
    response = requests.get(i)
    soup = BeautifulSoup(response.text,'html.parser')
    info = soup.find('div',class_='block-info container')
    address_list.append(info.find('a').text)
```

```python
# crawler the operation hours and ticketing hours of each station
operation_hours = []
ticketing_hours = []
for i in url_list:
    response = requests.get(i)
    soup = BeautifulSoup(response.text,'html.parser')
    info = soup.find('div',class_='block-info container')
    operation_hours.append(info.find_all('td',class_ = "gray")[0].string.split('：')[1])
    ticketing_hours.append(info.find_all('td',class_ = "gray")[1].string.split('：')[1])
```

```python
# create a DataFrame
ans = {'station':station_list,'address':address_list,'operation_hours':operation_hours,'ticketing_hours':ticketing_hours}
pd.DataFrame(ans)
```

# Crawler GDP and CPI
Let's start to crawl GDP and CPI!
* Target web: 
    1. [Wiki GDP](https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal))  
    2. [Wiki CPI](https://en.wikipedia.org/wiki/Corruption_Perceptions_Index)
* Task:
    1. crawler GDP table (top 50: United States ~ New Zealand)
    2. organize into a DataFrame (columns: Country, Region, IMF_Estimate, IMF_Year, United_Nations_Estimate, United_Nations_Year, World_Bank_Estimate, World_Bank_Year)
    3. crawler CPI table which contains country and 2020 CPI (top 100: Denmark ~ Suriname)
    4. organize into a DataFrame (columns: Country, CPI_2020)
    5. merge GDP(DataFrame) and CPI(DataFrame), based on Country of GDP
    6. plot and text the names of GDP top 10 countrys

```python
# crawler GDP
url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
Country = []
for i in soup.find_all(['tbody','td'],style='text-align:left'):
    Country.append(i.text.strip())
Country = Country[0:50]
Region = []
for i in soup.find_all(['tbody','td'],style='text-align:center'):
    Region.append(i.text)
Region = Region[0:50]
a = []
for i in soup.find_all(['tbody','td']):
    a.append(i.text)
b = a[8:408]
```

```python
# create GDP DataFrame
IMF_Estimate = []
IMF_Year = []
United_Nations_Estimate = []
United_Nations_Year = []
World_Bank_Estimate = []
World_Bank_Year = []
for i in range(len(b)):
    if i%8 == 2:
        IMF_Estimate.append(int(re.sub(r'[,]+', '', b[i])))    
for i in range(len(b)):
    if i%8 == 3:
        IMF_Year.append(b[i])
for i in range(len(b)):
    if i%8 == 4:
        United_Nations_Estimate.append(int(re.sub(r'[,]+', '', b[i])))    
for i in range(len(b)):
    if i%8 == 5:
        United_Nations_Year.append(b[i])
for i in range(len(b)):
    if i%8 == 6:
        World_Bank_Estimate.append(int(re.sub(r'[,]+', '', b[i])))       
for i in range(len(b)):
    if i%8 == 7:
        World_Bank_Year.append(int(b[i]))
IMF_Estimate = IMF_Estimate[0:50]
IMF_Year = IMF_Year[0:50]
United_Nations_Estimate = United_Nations_Estimate[0:50]
United_Nations_Year = United_Nations_Year[0:50]
World_Bank_Estimate = World_Bank_Estimate[0:50]
World_Bank_Year = World_Bank_Year[0:50]
ans = {'Country':Country,'Region':Region,'IMF_Estimate':IMF_Estimate,'IMF_Year':IMF_Year,
       'United_Nations_Estimate':United_Nations_Estimate,
       'United_Nations_Year':United_Nations_Year,
       'World_Bank_Estimate':World_Bank_Estimate,'World_Bank_Year':World_Bank_Year}
ans = pd.DataFrame(ans)
ans['IMF_Year'] = 2021
ans['United_Nations_Year'] = 2019
ans['World_Bank_Year'] = 2020
ans
```

```python
# crawler CPI
url = "https://en.wikipedia.org/wiki/Corruption_Perceptions_Index"
response = requests.get(url)
soup = BeautifulSoup(response.text,'html.parser')
soup.find_all('td',style='text-align:left;')
Country = []
for i in soup.find_all('td',style='text-align:left;'):
    Country.append(i.text.strip())
Country[0:100]
Country = Country[0:100]
soup.find_all('td')
c = []
for i in soup.find_all('td'):
    c.append(i.text)
d = c[23:1922]
CPI_2020 = []
for i in range(len(d)):
    if i%19 == 1:
        CPI_2020.append(int(d[i]))
```

```python
# CPI DataFrame
ans1 = {'Country':Country,'CPI_2020':CPI_2020}
ans1 = pd.DataFrame(ans1)
ans1
```

```python
# merge GDP and CPI DataFrame
ans2 = ans.merge(ans1)
ans2
```

```python
# plot
sns.scatterplot(x = ans2['CPI_2020'],y = ans2['World_Bank_Estimate'],hue = ans2['Region'],)
label = pd.concat({'CPI_2020': ans2['CPI_2020'][0:10], 'World_Bank_Estimate': ans2['World_Bank_Estimate'][0:10], 'Country':ans2['Country'][0:10]}, axis=1)
for i, point in label.iterrows():
    plt.gca().text(point['CPI_2020']+.9, point['World_Bank_Estimate'], str(point['Country']))
plt.legend(loc='right')
```
